```python
#Import modules and API key for Custom Search JSON API
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
harambe_file = "CSV_data/10years/harambe_file_y10.csv"
moth_file = "CSV_data/10years/moth_lammp_meme_file_y10.csv"
fox_file = "CSV_data/10years/what_does_the_fox_say_file_y10.csv"
cat_file = "CSV_data/10years/KeyboardCat_10y_file.csv"
fleek_file = "CSV_data/10years/on_fleek_file_y10.csv"
covfefe_file = "CSV_data/10years/covfefe_file_y10.csv"
rainbow_file = "CSV_data/10years/DoubleRainbow_y10_file.csv"
gangnamstyle_file = "CSV_data/10years/gangnamstyle_10y_file.csv"
rebeccablack_file = "CSV_data/10years/rebeccablackfriday_10y_file.csv"
whatwhat_file = "CSV_data/10years/WhatWhat_10y_file.csv"


harambe_file_df = pd.read_csv(harambe_file)
moth_file_df = pd.read_csv(moth_file)
fox_file_df = pd.read_csv(fox_file)
cat_file_df = pd.read_csv(cat_file)
fleek_file_df = pd.read_csv(fleek_file)
covfefe_file_df = pd.read_csv(covfefe_file)
rainbow_file_df = pd.read_csv(rainbow_file)
gangnamstyle_file_df = pd.read_csv(gangnamstyle_file)
rebeccablack_file_df = pd.read_csv(rebeccablack_file)
whatwhat_file_df = pd.read_csv(whatwhat_file)
```

```python
combined_memes10y_df = pd.merge(harambe_file_df, moth_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results_x': 'Harambe Meme', 'Results_y': 'Moth Lamp Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, fox_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Fox Say Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, cat_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Keyboard Cat Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, fleek_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'On Fleek Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, covfefe_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Covfefe Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, rainbow_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Dbl Rainbow Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, gangnamstyle_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Gangnam Style Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, rebeccablack_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Results': 'Rebecca Black Meme'}) 
combined_memes10y_df

combined_memes10y_df = pd.merge(combined_memes10y_df, whatwhat_file_df, how='outer', on='Months')
combined_memes10y_df
combined_memes10y_df = combined_memes10y_df.rename(columns= {'Months': 'Years Past', 'Results': 'What What Meme'}) 
combined_memes10y_df
```

```python
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
plt.plot(combined_memes10y_df["Harambe Meme"])
plt.plot(combined_memes10y_df["Moth Lamp Meme"])
plt.plot(combined_memes10y_df["Fox Say Meme"])
plt.plot(combined_memes10y_df["Keyboard Cat Meme"])
plt.plot(combined_memes10y_df["On Fleek Meme"])
plt.plot(combined_memes10y_df["Covfefe Meme"])
plt.plot(combined_memes10y_df["Dbl Rainbow Meme"])
plt.plot(combined_memes10y_df["Gangnam Style Meme"])
plt.plot(combined_memes10y_df["Rebecca Black Meme"])
plt.plot(combined_memes10y_df["What What Meme"])

plt.title("10 Year Visualization ALL")
plt.xlabel("Years")
plt.ylabel("Total Results")
plt.grid(alpha = 0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show
```
