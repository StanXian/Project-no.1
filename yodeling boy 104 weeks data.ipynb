```python
import json
import requests
import pandas as pd
from k import api_key
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
cx = "0004244157511163772131:swmjc3vgy_a"
url = "https://www.googleapis.com/customsearch/v1?"

total_results = []
weeks = []

#Yodeling boy
for x in range(105, 0, -1):
    print(x)
    timeperiod1 = 'w' + str(x)
    parameters1 = {"q": "yodelling walmart boy",
              "cx": cx,
              "key": api_key,
              "dateRestrict": timeperiod1
              }
    timeperiod2 = 'w' + str(x+1)
    parameters2 = {"q": "yodelling walmart boy",
              "cx": cx,
              "key": api_key,
              "dateRestrict": timeperiod2
              }
    
    #time.sleep(2)
    page1 = requests.request("GET", url, params=parameters1)
    page2 = requests.request("GET", url, params=parameters2)
    
    results1 = json.loads(page1.text)
    results2 = json.loads(page2.text)
    
    #print(results)
    hits1 = results1["searchInformation"]["totalResults"]
    hits1 = int(hits1)
    hits2 = results2["searchInformation"]["totalResults"]
    hits2 = int(hits2)
    
    #print(hits)
    hits_permon = hits2 - hits1
    total_results.append(hits_permon)
    weeks.append("W" + str(x))
```

```python
mon_df = pd.DataFrame({
    "Weeks": weeks,
    "Results": total_results
})
mon_df
```

```python
plt.plot(weeks, total_results)
plt.show
```

```python
mon_df.to_csv('yodelling walmart boy 104 weeks.csv', index = None, header=True) 
```
